```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('test.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
feature_nans=[feature for feature in df.columns  if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']

for feature in  feature_nans:
    print('{} :{} % Missing'.format(feature,np.round(df[feature].isnull().mean(),3)))
```

```python
## Replacing categorical feature with new feature:

def replacing_cat_feature(df,feature_nans):
    data=df.copy()
    data[feature_nans]=data[feature_nans].fillna('Missing')
    return data
df=replacing_cat_feature(df,feature_nans)
```

```python
df.head()
```

```python
df[feature_nans].isnull().sum()
```

```python
##  Numerical NAN values:

numerical_nans=[feature for feature in df.columns if df[feature].isnull().sum() >1 and df[feature].dtypes!='O']

for feature in numerical_nans:
    print('{} :{}%Missing' .format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
for feature in numerical_nans:
    
    median_value=df[feature].median()
    # Create a new feature to capture the nan values:
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace=True)
```

```python
df[numerical_nans].isnull().sum()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df[feature]=df['YrSold']-df[feature]
```

```python
df.head()
```

```python
df[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
#  Numerical Values are skewed:
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']

for feature in num_features:
    df[feature]=np.log(df[feature])
```

```python
df.head()
```

```python
# handling rare categorical values:

categorical_features=[feature for feature in df.columns if df[feature].dtypes=='O']
categorical_features
```

```python
for feature in  categorical_features:
    temp=df.groupby([feature])['SaleCondition'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
    
```

```python
df.head()
```

```python
df['SaleCondition'].unique()
```

```python
# replacing the categorical with valures:

for feature in categorical_features:
    label_order=df.groupby([feature])['YrSold'].mean().sort_values().index
    label_order={k:i for i,k in enumerate(label_order,0)}
    df[feature]=df[feature].map(label_order)
```

```python
df.head()
```

```python
# Feature Scaling:

feature_scaling=[feature for feature in df.columns if feature not in ['Id']]

from sklearn.preprocessing import MinMaxScaler

sc=MinMaxScaler()
sc.fit(df[feature_scaling])
```

```python
data=pd.concat([df[['Id']].reset_index(drop=True),
                  pd.DataFrame(sc.transform(df[feature_scaling]),columns=feature_scaling)],
                 axis=1)
```
